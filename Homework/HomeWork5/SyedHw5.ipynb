# Problem 2

```python
import xml.etree.ElementTree as ET

document = ET.parse('Homework 5 Input 2.mzXML')
root = document.getroot()


def count_spectra(root):
    return(len(root))

def scan_MS(root):
    count = 0
    for ele in root:
        if 'msLevel' in ele.attrib:
            if ele.attrib['msLevel'] == '2':
                count +=1
    return count

def count_precursor(root, low , high):
    count = 0
    for ele in root:
        if 'msLevel' in ele.attrib and ele.attrib['msLevel'] == '2':
            if ele.tag == '{http://sashimi.sourceforge.net/schema/}'+'precursorMz':
                if float(ele.text)>low and float(ele.text)<high:
                    count +=1
    return count

print('Count of spectra in the file:',count_spectra(root))
print('Count of MS/MS (attribute "msLevel" is 2) spectra (tag "scan") in the file:',scan_MS(root))
print('Count of MS/MS spectra have precursor m/z value between 600 and 920 Da: ',count_precursor(root, 600 , 920))
```

```python
from Bio import SeqIO
from Bio.Blast import NCBIWWW
import xml.etree.ElementTree as ET

query = SeqIO.read("Homework 5 Input 3.fasta",format="fasta")


result_handle = NCBIWWW.qblast('blastn','nt', query.seq)
blast_result = open('Homework 5 Output 3.xml','w')
blast_result.write(result_handle.read())
blast_result.close()
result_handle.close()


```

```python
import xml.etree.ElementTree as ET
import numpy as np
import pandas as pd

tree = ET.parse('Homework 5 Output 3.xml')
root = tree.getroot()

count =0
arr_index = []
e_val_arr = []

score_arr =[]
title_arr = []
len_title = []

f_score_arr = []
f_title_arr = []
f_len_title = []

for ele in root.iter('Hsp_evalue'):
    if float(ele.text) < 1e-50:
        e_val_arr.append(ele.text)
        arr_index.append(count)
    count +=1

#print(len(arr_index))   
#print(count)

for ele in root.iter('Hsp_score'):
    score_arr.append(ele.text)

for ele in root.iter('Hsp_qseq'):
    title_arr.append(ele.text[0:10])
    len_title.append(len(ele.text))
    

#print(len(e_val_arr))
#print(len(score_arr))
#print(len(title_arr))
#print(len(len_title))

for x in arr_index:
    f_score_arr.append(score_arr[x])
    f_title_arr.append(title_arr[x])
    f_len_title.append(len_title[x])
    
final_DataSet = list(zip(f_title_arr,f_len_title,f_score_arr,e_val_arr))
df = pd.DataFrame(data = final_DataSet,columns = ['Title','Length','Score','Expect'])
df.to_csv('FinalTable.csv', index = False)

df.head(100)
```
