```python
from tkinter import *
from tkinter.ttk import *
from Helper1 import symbolAndMass, atomCount, molesAndCompound
import re

window = Tk()
window.title("Chemical Equations HW 2")
window.geometry('900x300')

hash_periodic = symbolAndMass('PeriodicTableData.xlsx')

def function(avoi):
    outputcell1_r.insert(0,round(avoi[0],3))
    outputcell2_r.insert(0,round(avoi[1],3))
    outputcell3_r.insert(0,round(avoi[2],3))
    outputcell1_p.insert(0,round(avoi[3],3))
    outputcell2_p.insert(0,round(avoi[4],3))
    outputcell3_p.insert(0,round(avoi[5],3))

def moles_to_molecules(array_out_moles):
    array_out_molecules = []
    for x in range(6):
        array_out_molecules.append(array_out_moles[x]*6.022*(10**23))
    return array_out_molecules

def moles_from_GloMoles(moles_f,moles_h,Glo_moles):
    if(moles_f == 0):
        return 0
    else:
        return (Glo_moles * moles_f / moles_h)
    

def moles_out(Moles,User_pos_data):
    Arrayout = []
    for x in range(6):
        if(User_pos_data[0] == x):
            Arrayout.append(0)
        else:
            Arrayout.append(moles_from_GloMoles(Moles[x],Moles[User_pos_data[0]],User_pos_data[3]))
    return Arrayout

def get_moles_pos(User_pos_data):
    moles = User_pos_data[1]/User_pos_data[2]
    return moles
    

def pos_and_value(AMU_COMP_arr):
    Array_weight = [outputcell1_r.get(),outputcell2_r.get(),outputcell3_r.get(),outputcell1_p.get(),outputcell2_p.get(),outputcell3_p.get()]
    for x in range(6):
        if(Array_weight[x] != ""):
            return [x,float(Array_weight[x]),AMU_COMP_arr[x],0] #has moles at last index
    
def MolarMass_comp(moles_arr,AMU_array):
    Molar_mass = []
    for x in range(6):
        Molar_mass.append(moles_arr[x]*AMU_array[x])
    return Molar_mass
    
def Compound_AMU(comp_text_pure):
    AMU = 0.0 
    if(comp_text_pure == 0):
        return 0.0
    else:
        hash_element  = atomCount(comp_text_pure)
        for key,val in hash_element.items():
            AMU = AMU + (val*float(hash_periodic[key]))
    return AMU


def get_AMU_ARRAY(arr_comp_PURE):
    AMU_arr = []
    for x in range(6):
        AMU_arr.append(Compound_AMU(arr_comp_PURE[x]))
    return AMU_arr
        
    
def calculate_M_compound(Compound_text):
    if(Compound_text == 0):
        return [0,0]
    else:
        return molesAndCompound(Compound_text)
    
def moles_from_text(Array_text):
    Data,Moles= [],[]
    for x in range(6):
        Data.append(calculate_M_compound(Array_text[x]))
        Moles.append(Data[x][0])
    return Moles
    
def CompPure_from_text(Array_text):
    Data,CompPure= [],[]
    for x in range(6):
        Data.append(calculate_M_compound(Array_text[x]))
        CompPure.append(Data[x][1])
    return CompPure
    
def make_eff_array(non_eff_Array):
        for x in range(6):
            if(non_eff_Array[x] == ''):
                non_eff_Array[x] = 0
        return non_eff_Array
    
def get_comp_text():
    Array_compunds_in = [inputcell1_r.get(),inputcell2_r.get(),inputcell3_r.get(),inputcell1_p.get(),inputcell2_p.get(),inputcell3_p.get()]
    Array_compunds_in = make_eff_array(Array_compunds_in)
    return Array_compunds_in


    
def calculate():
    Arr_Text_comp = get_comp_text()
    Moles = moles_from_text(Arr_Text_comp)
    Comp_Pure = CompPure_from_text(Arr_Text_comp)
    AMU_PURE_arr = get_AMU_ARRAY(Comp_Pure)
    AMU_COMP_arr = MolarMass_comp(Moles,AMU_PURE_arr)
    User_pos_data = []
    User_pos_data= pos_and_value(AMU_COMP_arr)
    Glo_moles = get_moles_pos(User_pos_data)
    User_pos_data[3] = Glo_moles
    #added moles of output to User ouput data
    moles_fill_out_ARR = moles_out(Moles,User_pos_data)
    if(molecules.get() == 1):
        Molecules_fill_out_ARR = moles_to_molecules(moles_fill_out_ARR)
        function(Molecules_fill_out_ARR)
    else:
        function(moles_fill_out_ARR)
    
    
    
    
#Text (0,0)
First_text = Label(window, text = "Enter Balanced Equation: ")
First_text.grid(column = 0, row = 0)

#Spacer between the two boxes Vertical
#spacer = Label(window, text = " ")
#spacer.grid(column = 0, row = 1)

#Button to run program
btnCell = Button(window, text="Compute", command=calculate)
btnCell.grid(column = 0, row=2, sticky = 'W')

#Check button to convert to molecules
#Clicking the check button makes the molecules variable 1
molecules = IntVar()
Checkbutton(window, text = "In Molecules", variable = molecules).grid(column = 0,row = 1, sticky = 'W')


#First row of Input Boxes
inputcell1_r= Entry(window, width = 15)
inputcell1_r.grid(column = 1, row = 0)

plussign = Label(window,text = " + ")
plussign.grid(column = 2, row = 0)

inputcell2_r= Entry(window, width = 15)
inputcell2_r.grid(column = 3, row = 0)

plussign = Label(window,text = " + ")
plussign.grid(column = 4, row = 0)

inputcell3_r= Entry(window, width = 15)
inputcell3_r.grid(column = 5, row = 0)

equalsign = Label(window,text = " = ")
equalsign.grid(column = 6, row = 0)

inputcell1_p= Entry(window, width = 15)
inputcell1_p.grid(column = 7, row = 0)

plussign = Label(window,text = " + ")
plussign.grid(column = 8, row = 0)

inputcell2_p= Entry(window, width = 15)
inputcell2_p.grid(column = 9, row = 0)

plussign = Label(window,text = " + ")
plussign.grid(column = 10, row = 0)

inputcell3_p= Entry(window, width = 15)
inputcell3_p.grid(column = 11, row = 0)



#Second row of Output, 1 input of grams
outputcell1_r= Entry(window, width = 15)
outputcell1_r.grid(column = 1, row = 2)

outputcell2_r= Entry(window, width = 15)
outputcell2_r.grid(column = 3, row = 2)

outputcell3_r= Entry(window, width = 15)
outputcell3_r.grid(column = 5, row = 2)

equalsign = Label(window,text = " = ")
equalsign.grid(column = 6, row = 2)

outputcell1_p= Entry(window, width = 15)
outputcell1_p.grid(column = 7, row = 2)

outputcell2_p= Entry(window, width = 15)
outputcell2_p.grid(column = 9, row = 2)

outputcell3_p= Entry(window, width = 15)
outputcell3_p.grid(column = 11, row = 2)



window.mainloop()
```
